```python
# Convert manually labeled bond Excel data to train_bonds.jsonl
import pandas as pd
import json

# Load Excel file (update with your actual path)
df = pd.read_excel("/home/harish/Project_works/zResearch/DataMiningResultsCorporateAndGovernment.xlsx")

# Define relevant fields to include in JSON output
fields_to_extract = [
    "Name", "ISIN", "FRNBenchmark", "Margin", "Seniority", "Putoption", "putdate",
    "putamount", "putfrequency", "OriginalAmountissued", "Couponfrequency", "Calloption",
    "Issuancecoupon", "Bondtype", "Coupontype", "Issuedate", "Maturitydate",
    "Issuecurrency", "Finalredemptionvalue", "ParCall", "Clean up Call", "calleddate",
    "callamount", "callfrequency", "greenbond", "Perpetual", "Comments"
]

# Drop rows with missing ISIN or no relevant output fields
df = df[df["ISIN"].notnull()]

# Fill NaN with None for clean JSON output
df = df.where(pd.notnull(df), None)

# Create the training data format
jsonl_data = []

for idx, row in df.iterrows():
    output_dict = {field: row[field] for field in fields_to_extract if field in row}
    jsonl_data.append({
        "instruction": "Extract bond details.",
        "input": f"Bond ISIN: {row['ISIN']}, Issuer: {row['Name']}, Maturity: {row['Maturitydate']}.",
        "output": output_dict
    })

# Write to JSONL file
with open("train_bonds.jsonl", "w") as f:
    for item in jsonl_data:
        f.write(json.dumps(item) + "\n")

print(f"‚úÖ Converted {len(jsonl_data)} rows to train_bonds.jsonl")
```
